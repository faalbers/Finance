```python
import matplotlib.pyplot as plt
# for jupyter
# %matplotlib inline
import numpy as np
import pandas as pd
```

```python
x_1 = np.linspace(0, 5, 20)
y_1 = x_1**2
plt.plot(x_1, y_1)
plt.title('Square Chart')
plt.xlabel('in')
plt.ylabel('out')
# plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x_1, y_1, 'r')
plt.subplot(1,2,2)
plt.plot(x_1, y_1, 'b')
```

```python
fig_1 = plt.figure(figsize=(5,4), dpi=100)
axes_1 = fig_1.add_axes([0.1, 0.1, 0.9, 0.9])
axes_1.set_xlabel('in')
axes_1.set_ylabel('out')
axes_1.set_title('Big one')
axes_1.plot(x_1, y_1, label='X/X^2')
axes_1.plot(y_1, x_1, label='X^2/X')
# loc: 0 = best spot, 1 = upper right, 2 = upper left, 3 = lower left, 4 = lower right
# or tuple (x,y) from lower left
axes_1.legend(loc=0)

axes_2 = fig_1.add_axes([0.45, 0.45, 0.4, 0.3])
axes_2.set_xlabel('in')
axes_2.set_ylabel('out')
axes_2.set_title('Small one')
axes_2.plot(x_1, y_1, 'r')
axes_2.text(1, 15, 'Text line')
```

```python
fig_2, axes_2 = plt.subplots(figsize=(8,5), nrows=1, ncols=3)
plt.tight_layout()
axes_2[1].set_title('Plot 2')
axes_2[1].set_xlabel('X')
axes_2[1].set_ylabel('X Squared')
axes_2[1].plot(x_1, y_1)
```
